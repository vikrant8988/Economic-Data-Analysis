```python
world_bank = pd.read_excel("WorldBank.xlsx").rename(columns={'Country Name' : "Country"})
```

```python
world_bank.head(5)
```

```python
world_bank["Population(M)"] = (world_bank["GDP (USD)"]/world_bank["GDP per capita (USD)"] / 1_000_000) 
```

```python
world_bank.head(5)
```

```python
world_bank.head(5)
```

```python
world_bank.info()
```

```python
world_bank.describe()
```

```python
hdi = pd.read_csv("HDI.csv")
```

```python
hdi.head()
```

```python
wb_hdi_2014 = world_bank.query("Year == 2014").merge(hdi[["iso3","hdi_2014"]], how="left", left_on="Country Code", right_on="iso3" )
```

```python
wb_hdi_2014.head()
```

```python
wb_hdi_2014.describe()
```

```python
gdp_pivot = world_bank.pivot_table(index="Year",columns="Region",values="GDP (USD)",aggfunc="sum")
gdp_pivot.head()
```

```python
pop_pivot = world_bank.pivot_table(index="Year",columns="Region",values="Population(M)",aggfunc="sum")
pop_pivot.head()
```

```python
wb_hdi_by_region = wb_hdi_2014.groupby("Region").agg({"hdi_2014" :"mean"}).sort_values("hdi_2014",ascending=False)
wb_hdi_by_region.head()
```

```python
gdp_pivot.iloc[-1].sort_values(ascending=False)
```

```python
gdp_pivot.tail()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig,ax = plt.subplots()

ax.stackplot(
    gdp_pivot.index,
    [gdp_pivot[region]/1_000_000_000_000 for region in gdp_pivot.iloc[-1].sort_values(ascending=False).index],
    labels=gdp_pivot.iloc[-1].sort_values(ascending=False).index
)

ax.legend(loc="upper left")
ax.set_title("GDP has grown exponentially over the Time")
ax.set_ylabel("GDP (Trillions)")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig,ax = plt.subplots()

ax.stackplot(
    pop_pivot.index,
    [pop_pivot[region]/1000 for region in pop_pivot.iloc[-1].sort_values(ascending=False).index],
    labels=pop_pivot.iloc[-1].sort_values(ascending=False).index
)

ax.legend(loc="upper left")
ax.set_title("Population has increased from 2 Billion to 7.5 Billion")
ax.set_ylabel("Population (Billion)")
```

```python
minsize = min(wb_hdi_2014["Population(M)"])
maxsize = max(wb_hdi_2014["Population(M)"])

sns.scatterplot(
    data=wb_hdi_2014,
    x="Life expectancy at birth (years)",
    y="GDP per capita (USD)",
    size="Population(M)",
    sizes=(minsize,maxsize),
    hue="Region",
    legend=False
).set(yscale="log",title="Life Expectancy increases as country gets richer")
```

```python
import matplotlib.gridspec as gridspec
import matplotlib as mtl

fig = plt.figure(figsize=(8,12))

gs = gridspec.GridSpec(nrows=12,ncols=6)
mtl.rcParams["font.family"]="Consolas"

fig.suptitle(
    "Report Global Economic Development from 1960-2018",
    fontsize=16,
    fontweight="bold",
    x=.65,
    y=.93
)

stack1_list = ['#66c2a5','#fc8d62','#8da0cb','#e78ac3','#a6d854','#ffd92f','#b3b3b3']
stack2_list = ['#66c2a5','#a6d854','#b3b3b3','#fc8d62','#e78ac3','#8da0cb','#ffd92f']
bubble_list=['#a6d854','#fc8d62','#ffd92f','#66c2a5','#b3b3b3','#e78ac3','#8da0cb']
bar_list =['#8da0cb','#fc8d62','#e78ac3','#ffd92f','#66c2a5','#a6d854','#b3b3b3']

ax1 = fig.add_subplot(gs[0:4,0:3])
ax1.stackplot(
    gdp_pivot.index,
    [gdp_pivot[region]/1_000_000_000_000 for region in gdp_pivot.iloc[-1].sort_values(ascending=False).index],
    labels=gdp_pivot.iloc[-1].sort_values(ascending=False).index,
    colors=stack1_list
)
ax1.set_title("GDP has grown exponentially over the Time",fontdict={'fontsize': 8.5, 'fontweight': 'bold'})
ax1.set_ylabel("GDP (Trillions)")
fig.text(
    0.92,
    0.67,
    ''' Over the years, GDP has seen rapid exponential growth, 
    meaning it's been increasing at an accelerating rate. Meanwhile, 
    population growth has been steady and
    linear, adding roughly the same number of people each year. 
    This combination implies that the economy’s productivity 
    per person has been rising significantly.''',
    {'fontsize':12}
)


ax2 = fig.add_subplot(gs[0:4,3:6])
ax2.stackplot(
    pop_pivot.index,
    [pop_pivot[region]/1000 for region in pop_pivot.iloc[-1].sort_values(ascending=False).index],
    labels=pop_pivot.iloc[-1].sort_values(ascending=False).index,
    colors=stack2_list
)

ax2.legend(bbox_to_anchor=(1.05,1),fontsize=8)
ax2.set_title("Population has increased from 2 Billion to 7.5 Billion",fontdict={'fontsize': 8.5, 'fontweight': 'bold'})
ax2.set_ylabel("Population (Billion)")

fig.text(
    0.92,
    0.45,
    '''By visualizing, we can see that life expectancy 
    and GDP per capita have a linear relationship, 
    meaning as GDP per capita increases, 
    life expectancy tends to rise at a steady rate..
    India and China holds large population and
    has potential of high GDP Country''',
    {'fontsize':12}
)


ax3 = fig.add_subplot(gs[4:8,:])
minsize = min(wb_hdi_2014["Population(M)"])
maxsize = max(wb_hdi_2014["Population(M)"])

sns.scatterplot(
    data=wb_hdi_2014,
    x="Life expectancy at birth (years)",
    y="GDP per capita (USD)",
    size="Population(M)",
    sizes=(minsize,maxsize),
    hue="Region",
    legend=True,
    ax=ax3,
    palette=bubble_list
).set(yscale="log",title="Life Expectancy increases as country gets richer")
handles,labels=ax3.get_legend_handles_labels()
enteries_to_skip=len(wb_hdi_2014["Region"].unique())+1
ax3.legend(
    handles[enteries_to_skip:],
    labels[enteries_to_skip:],
    bbox_to_anchor=(.6,.9),
    borderaxespad=0,
    frameon=False,
    ncols=6,
    fontsize=6
)

ax4 = fig.add_subplot(gs[8:12,0:3])
ax4.bar(wb_hdi_by_region.index,
        wb_hdi_by_region["hdi_2014"].values,
        color=bar_list)
ax4.set_title("HDI by region")
ax4.set_xticks(ax4.get_xticks(), ax4.get_xticklabels(), rotation=90, ha='right')
ax4.set_ylabel("Human Development Index (HDI)")


ax5 = fig.add_subplot(gs[8:12,3:6])
sns.scatterplot(
    data=wb_hdi_2014.query("Country != 'Iceland'"),
    x='Electric power consumption (kWh per capita)',
    y= 'GDP per capita (USD)',
    hue='hdi_2014',
    palette='coolwarm_r',
    ax=ax5
).set(title="Electricty Drives Development")

fig.text(
    0.92,
    0.15,
    '''States with high GDP tend to have higher electricity consumption, 
    indicating that as a state's economy grows, 
    its demand for electricity will also rise. This correlation suggests 
    that economic expansion often leads to increased industrial activity,
    higher energy needs for businesses, and greater residential consumption. 
    Consequently, ensuring a reliable and adequate electricity supply becomes crucial 
    for sustaining growth and meeting the energy demands of a growing population and economy.''',
    {'fontsize':12}
)


fig.subplots_adjust(hspace=4, wspace=4)

plt.show()
```
